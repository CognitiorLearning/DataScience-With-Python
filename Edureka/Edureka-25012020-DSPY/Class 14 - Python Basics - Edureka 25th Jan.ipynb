```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Recommend.csv', names = ['user_id','movie_id','rating','timestamp'])
df
```

```python
from sklearn.model_selection import train_test_split
n_users = df.user_id.unique().shape[0]
```

```python
n_users
```

```python
n_movies = df.movie_id.unique().shape[0]
```

```python
n_movies
```

```python
train_data, test_data = train_test_split(df, test_size = 0.25)
```

```python
train_data_matrix = np.zeros((n_users, n_movies))
```

```python
for line in train_data.itertuples():
    train_data_matrix[line[1]-1, line[2]-1] = line[3]
pd.DataFrame(train_data_matrix)
```

```python
test_data_matrix = np.zeros((n_users, n_movies))
for line in test_data.itertuples():
    test_data_matrix[line[1]-1, line[2]-1] = line[3]
pd.DataFrame(test_data_matrix)
```

```python
from sklearn.metrics import pairwise_distances
```

```python
user_similarity = pairwise_distances(train_data_matrix, metric = 'cosine')
mean_user_rating = train_data_matrix.mean(axis=1)[:, np.newaxis]
rating_diff = (train_data_matrix-mean_user_rating)
user_pred = mean_user_rating + user_similarity.dot(rating_diff) /np.array([np.abs(user_similarity).sum(axis=1)]).T
pd.DataFrame(user_pred)
```
