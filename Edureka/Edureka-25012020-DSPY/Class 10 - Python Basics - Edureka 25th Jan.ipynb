```python
from mlxtend.frequent_patterns import apriori
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('Online_Retail.xlsx')
```

```python
len(df)
```

```python
df.info()
```

```python
df['Description'] = df['Description'].str.strip()
```

```python
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
```

```python
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
basket = (df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket.to_csv('output.csv')
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
basket_sets
```

```python
frequent_items = apriori(basket_sets, min_support = 0.07, use_colnames=True)
rules = association_rules(frequent_items, metric='lift', min_threshold=1)
len(rules)
```

```python
rules
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:,[2,3]].values
```

```python
x
```

```python
y = dataset.iloc[:,4].values
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x= sc.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', C=6)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
y_pred
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors=5)
classifier_knn.fit(x_train, y_train)
```

```python
classifier_knn.predict(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'C':[2,3], 'kernel':['rbf','sigmoid']}]
grid_search = GridSearchCV(estimator=classifier,
                          param_grid=parameters,
                          scoring='accuracy',
                          n_jobs=-1)
grid_search = grid_search.fit(x_train, y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'n_neighbors':[43,45,47]}]
grid_search = GridSearchCV(estimator=classifier_knn,
                          param_grid=parameters,
                          scoring='accuracy',
                          n_jobs=-1)
grid_search = grid_search.fit(x_train, y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
#10-fold cross validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X=x_train, y=y_train, cv=10)
```

```python
accuracies.mean()
```
