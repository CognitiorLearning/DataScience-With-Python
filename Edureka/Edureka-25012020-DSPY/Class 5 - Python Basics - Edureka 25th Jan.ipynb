```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
```

```python
year = [2015, 2016, 2017, 2018, 2019]
sales = [3224,3455,3433,4545,4434]
profit = [324,544,555,344,333]
```

```python
plt.plot(year,sales,'red')
plt.plot(year, profit, 'blue')
plt.title('Sales and Profit by Year')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(year, sales, 'red')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Sales by Year')

plt.subplot(1,2,2)
plt.plot(year, profit, 'blue')
plt.xlabel('Years')
plt.ylabel('profit')
plt.title('profit by Year')
plt.tight_layout()
plt.show()
```

```python
plt.plot(year, sales , 'red')
#plt.xlim((2016,2018))
#plt.ylim((3500, 4000))
plt.axis((2016,2018, 3500,4000))
```
