```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
x=dataset.iloc[:,2:4].values
```

```python
x
```

```python
y=dataset.iloc[:,4].values
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x=sc_x.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
```

```python
y_pred_log=logmodel.predict(x_test)
```

```python
y_pred_log
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier()
classifier_knn.fit(x_train, y_train)
```

```python
y_pred_knn=classifier_knn.predict(x_test)
```

```python
sum(y_pred_knn)
```

```python
y_test
```

```python
sum(y_pred_knn)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_log)
```

```python
confusion_matrix(y_test, y_pred_knn)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(x_train, y_train)
```

```python
classifier_nb.predict(x_test)
```

```python
from sklearn.svm import SVC
classifier_svm_lin = SVC(kernel='linear')
classifier_svm_lin.fit(x_train,y_train)
```

```python
classifier_svm_lin.predict(x_test)
```

```python
from sklearn.svm import SVC
classifier_svm_rbf = SVC(kernel='rbf')
classifier_svm_rbf.fit(x_train,y_train)
```

```python
classifier_svm_rbf.predict(x_test)
```

```python
from sklearn.svm import SVC
classifier_svm_sig = SVC(kernel='sigmoid')
classifier_svm_sig.fit(x_train,y_train)
```

```python
y_pred_svm_sig=classifier_svm_sig.predict(x_test)
```

```python
confusion_matrix(y_test, y_pred_svm_sig)
```

```python
from sklearn.svm import SVC
classifier_svm_poly = SVC(kernel='poly')
classifier_svm_poly.fit(x_train,y_train)
```
