```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('Online_Retail.xlsx')
```

```python
df['Description'] = df['Description'].str.strip()
```

```python
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
```

```python
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
df
```

```python
basket = (df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket.to_csv('output1.csv')
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.to_csv('output2.csv')
```

```python
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
len(rules)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
y = dataset.iloc[:,4].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='sigmoid')
classifier.fit(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier()
classifier_knn.fit(x_train, y_train)
```

```python
y_pred_svm = classifier.predict(x_test)
```

```python
y_pred_knn = classifier_knn.predict(x_test)
```

```python
y_pred_svm
```

```python
y_pred_knn
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred_svm)
accuracy_score(y_test, y_pred_svm)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred_knn)
accuracy_score(y_test, y_pred_knn)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'C':[1,2,3,4], 'kernel':['sigmoid','rbf','linear']}]
grid_search = GridSearchCV(estimator=classifier, 
                          param_grid=parameters,
                          scoring='accuracy',
                          n_jobs=-1)
grid_search = grid_search.fit(x_train, y_train)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
```
