```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```python
transactions = []
for i in range(0,7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
transactions
```

```python
from apyori import apriori
rules = apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift =6, min_length=2)
```

```python
results = list(rules)
```

```python
pd.DataFrame(results)
```
